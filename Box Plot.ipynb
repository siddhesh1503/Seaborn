```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",order=["Fri","Sun","Thur","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(data=var,showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},orient="h")
plt.show()
```

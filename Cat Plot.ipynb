```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips").head(20)
var
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette="Oranges")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",kind="point")
plt.show()
```

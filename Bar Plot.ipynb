Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm,data=var)
plt.show()
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm,data=var,hue="sex")
plt.show()
```

```python
#Order parameter
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x=var.island,y=var.bill_length_mm,data=var,hue="sex",order=order_1,hue_order=["Female","Male"])
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x=var.island,y=var.bill_length_mm,data=var,hue="sex",order=order_1,hue_order=["Female","Male"],errorbar=('ci', 100),n_boot=2)
plt.show()
```

```python
#For the horizantal
sns.barplot(x=var.island,y=var.bill_length_mm,data=var,orient="h")
plt.show()
```

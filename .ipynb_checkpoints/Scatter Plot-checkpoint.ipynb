```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                data=var,
                hue="sex",
                style="sex",
                sizes=(40, 200))  # Min and max size range
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                data=var,
                hue="sex",
                style="sex",
                sizes=(40, 200),  # Min and max size range
                palette="gist_ncar")  # Color palette
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                data=var,
                hue="sex",
                style="sex",
                sizes=(40, 200),  # Min and max size range
                palette="gist_ncar",
               alpha=1)  # Color palette
plt.show()
```

```python
n={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                data=var,
                hue="sex",
                style="sex",
                sizes=(40, 200),  # Min and max size range
                palette="gist_ncar",
                markers=n)  # Color palette
plt.show()
```

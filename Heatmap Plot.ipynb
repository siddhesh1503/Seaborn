```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var1 = np.linspace(1,10,20).reshape(4,5)
var1
```

```python
sns.heatmap(var1,vmin=0,vmax=20)
plt.show()
```

```python
var = sns.load_dataset("anagrams")
x = var.drop(columns=["attnr"],axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(var1,vmin=0,vmax=20,cmap="PuOr",annot=True)
plt.show()
```

```python
var2 = np.linspace(1,10,10).reshape(2,5)
var2
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
              ["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var2,vmin=0,vmax=20,cmap="PuOr",annot=ar,fmt="s")
plt.show()
```
